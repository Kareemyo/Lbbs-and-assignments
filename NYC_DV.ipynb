```python
os.chdir(r"/Users/arnel/Downloads") 
data = pd.read_csv("Large_NYC_Visualization_Dataset.csv") #Read the file into a dataframe
data.head() # Preview the first 5 rows of df
```

```python
# Create figure and set plot size
plt.figure(figsize = (22, 8))

# Generate plot with attributes
sns.kdeplot(data, x = "Unemployment_Rate", hue = "Neighborhood", fill = False)
plt.title("Unemployment Rate by Neighborhood")
plt.xlabel("Unemploment Rate")
plt.ylabel("Density")
plt.show()
```

The reason for me choosing this visualization because I wanted to see the distribution  of the unemployment rate within the neighborhoods  but given that the rate was so varied, I thought looking at the density curves  instead would provide a more smooth transition as it uses a Gaussian kernel which produces an estimate. I observe by doing this that the neighborhood that faces the highest unemployment rates is Brooklyn Although not being the most popululated. In addition to Queens being the neighborhood being accounting for the lowest unemployment rates.

```python
plt.figure(figsize=(12, 6))  # Set the figure size
sns.countplot(data=data, x='Neighborhood', hue='Event', palette='Set2')

# Add labels and title
plt.title('Events by Neighborhood', fontsize=16)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Event Count', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

The reason why I chose this visualization because I wanted  to view a side by side comparison of each Unique value in the Events column based on the Neighborhood. From this i was able to deduce that most festivals were held in Brooklyn, given the number of large parks it has. Also I can see that the neighborhood of Manhattan  has the most sporting events and parade which comes as no surprise since it home to Madison Sq gardens along with very popilar parades like the Macy's Thanksgiving Day Parade, LGBTQ etc. Each neighborhood were prone the roughly the same amount of protests.
